```python
from skimage.data import coins
import matplotlib.pyplot as plt
from skimage.transform import hough_circle_peaks, hough_circle
from skimage.filters import sobel
import numpy as np
from skimage.morphology import binary_opening, binary_closing
from skimage import color
from skimage.draw import circle_perimeter
from skimage import io
```

```python
img = 'images/105549.jpg'
imagem_original = io.imread(img)

plt.title('Imagem Original')
plt.imshow(imagem_original, cmap = 'gray') #carregar a imagem usando a biblioteca para poder chamar o metodo copy na str img
plt.axis('off')  # Desativa os eixos
plt.show()
```

```python
imagem_cinza = color.rgb2gray(imagem_original)
plt.imshow(imagem_cinza, cmap='gray')
```

```python
histogram = np.histogram(imagem_cinza.ravel(), bins=256)
plt.plot(histogram[0], '-b')
```

```python
imagem_binaria=imagem_cinza.copy()
limiar = imagem_cinza.max()*(110/256)
imagem_binaria[imagem_binaria<=limiar]=0
imagem_binaria[imagem_binaria>0]=1
plt.imshow(imagem_binaria*255,cmap='gray')
```

```python
imagem_binaria = binary_opening(imagem_binaria)
imagem_binaria = binary_closing(imagem_binaria)
plt.imshow(imagem_binaria, cmap='gray')
```

```python
bordas = sobel (imagem_binaria)
plt.imshow(bordas, cmap='gray')
```
