```python
from skimage.data import coins
import matplotlib.pyplot as plt
from skimage.transform import hough_circle_peaks, hough_circle
from skimage.filters import sobel
import numpy as np
from skimage.morphology import binary_opening, binary_closing
from skimage import color
from skimage.draw import circle_perimeter
from skimage import io
```

```python
img = '100.jpg'
imagem_original = io.imread(img)

plt.title('Imagem Original')
plt.imshow(imagem_original, cmap = 'gray') #carregar a imagem usando a biblioteca para poder chamar o metodo copy na str img
plt.axis('off')  # Desativa os eixos
plt.show()
```

### Histograma da imagem

```python
histogram = np.histogram(imagem_original.ravel(), bins=256)
```

```python
plt.plot(histogram[0], '-b')
```

### Binarizando a imagem

```python
imagem_binaria = imagem_original.copy()
limiar = imagem_original.max()*(110/256)
imagem_binaria[imagem_binaria<=limiar]=0
imagem_binaria[imagem_binaria>0]=1
plt.imshow(imagem_binaria*255, cmap='gray')
```
